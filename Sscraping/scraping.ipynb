```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

# Open browser
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Open a website
driver.get("https://www.flipkart.com")

print(driver.title)
```

```python
from selenium.webdriver.common.keys import Keys


search_box = driver.find_element(By.NAME, "q")
search_box.send_keys("laptops")
search_box.send_keys(Keys.RETURN)
```

```python
import time
time.sleep(2)  # wait for page to load

products = driver.find_elements(By.CLASS_NAME, "RG5Slk")
for product in products[:5]:
    print(product.text)
```

```python
import time
time.sleep(2) 

products = driver.find_elements(By.CSS_SELECTOR, ".RG5Slk")
prices = driver.find_elements(By.CLASS_NAME, "hZ3P6w")

for product, price in zip(products[:5], prices[:5]):
     print(f"{product.text} - {price.text}")
```

```python
import pandas as pd

df = pd.DataFrame({
    'product': [product.text for product in products[:5]],
    'price': [price.text for price in prices[:5]]
})
df.to_csv('laptops.csv', index=False)
```
