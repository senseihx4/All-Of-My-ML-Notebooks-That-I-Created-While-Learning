```python
import numpy as np
import matplotlib.pyplot as plt

# House sizes (sq ft)
X = np.array([800, 1100, 1300, 1500, 1700, 1900, 2100, 2300, 2500, 2800])

# Prices ($k)
y = np.array([150, 195, 210, 260, 295, 320, 355, 370, 410, 460])
```

```python
def predict(x,w,b):
    return w * x + b


predictions = predict(X, w=0.15, b=50)
print(predictions)
np.mean(predictions)
np.sum(predictions)
```

```python
# Task 2 — three different lines
lines = [
    (0.05, 100, "flat line, high base"),
    (0.30, 10,  "steep line, low base"),
    (0.15, 30,  "reasonable guess"),
]

plt.figure(figsize=(9, 5))
plt.scatter(X, y, color='purple', zorder=5, label='Actual prices', s=60)

colors = ['red', 'orange', 'green']
for (w, b, label), color in zip(lines, colors):
    y_pred = predict(X, w, b)
    plt.plot(X, y_pred, color=color, label=f'w={w}, b={b} ({label})')

plt.xlabel('House size (sq ft)')
plt.ylabel('Price ($k)')
plt.title('Linear regression — trying different w and b')
plt.legend()
plt.tight_layout()
plt.show()
```
